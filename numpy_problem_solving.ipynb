```python
print("---Reshape the given List---")
import numpy as np
products=np.array([134,453,354,754,643,747])
reshape=products.reshape(2,3)
print(reshape)
```

```python
print("---use flatten & ravel---")
val=np.array([[547,743,637,737],[563,365,653,753]])
flat=val.flatten()
print(flat)
print("change the value")
flatt=val.flatten()
flatt[0]=638
print(flatt)
print("check orginal value")
print(flat)
print("using ravel")
ravel=val.ravel()
print(ravel)
print("change the value")
r=val.ravel()
r[1]=72
print(r)
print("check if orginal value changed permentally")
print(ravel)
```

```python
print("---1D Arithmetic operations using numpy---")
val1 = np.array([120, 150, 180])
val2 = np.array([100, 140, 160])
print(val1+val2)
print(val1-val2)
print(val1*val2)
print(val1/val2)
print(val1**2)
```

```python
print("---2D Arithmetic operations using numpy---")
value1=np.array([[43,63,83,51],[53,78,36,74]])
value2=np.array([47,67,74,12])
print("\nadd",value1+value2)
print('\nsub',value1-value2)
print('\nmul',value1*value2)
print('\ndiv',value1/value2)
```

```python
print("--- add column-wise @ row-wise---")
print(value1.sum(axis=0))
print(value1.sum(axis=1))
```
