```python
print("Using Class Variable\n")
class College:
    facilities = "WiFi, Library, Smart Board"
    def cse(self):
        print("CSE students can use:", College.facilities)
    def eee(self):
        print("EEE students can use:", College.facilities)
c=College()
c.cse()
c.eee()
    
```

```python
print("Using Both Class & Local Variable\n")
class College:
    facilities = "WiFi, Library, Smart Board"
    def cse(self):
        play="Cricket Ground"
        print("CSE students can use:", play)
        print("CSE students can use:", College.facilities)
    def eee(self):
        print("EEE students can use:", College.facilities)
c=College()
c.cse()
c.eee()
```

```python
print("Using Global Variable\n")
clg='KSR College of Technology'
class College:
    def cse(self):
        print("CSE students from: ", clg)
c=College()
c.cse()
```
