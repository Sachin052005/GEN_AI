```python
print("---Cab Booking System---")
from abc import ABC, abstractmethod
class cab(ABC):
    @abstractmethod
    def fare(self,km):
        pass
class bike(cab):
    def fare(self,km):
        return km*5
class mini(cab):
    def fare(self,km):
        return km*10
class sedan(cab):
    def fare(self,km):
        return km*13
class Ride:
    def __init__(self,customer,cab,distance):
        self.customer=customer
        self.cab=cab
        self.distance=distance

    def bill(self):
        return self.cab.fare(self.distance)

rides=[
      Ride('Sachin',sedan(),30),
      Ride('Joshua',mini(),25),
      Ride('Infant',bike(),26)]

for c in rides:
    print(c.customer,'Paid $: ',c.bill())
```

```python
print("---Notification System---")
class Notification:
    def send(self, msg):
        pass
class Email(Notification):
    def send(self, msg):
        print("Email:", msg)
class SMS(Notification):
    def send(self, msg):
        print("SMS:", msg)
class Push(Notification):
    def send(self, msg):
        print("Push:", msg)
notifications = [Email(), SMS(), Push()]
for n in notifications:
    n.send("Order Delivered Successfully")
```
