```python
subjects = ['Java', 'Python', 'AI', 'C', 'DBMS']
java = int(input('Enter Java mark:'))
python = int(input('Enter Python mark:'))
ai = int(input('Enter AI mark:'))
c = int(input('Enter C mark:'))
dbms = int(input('Enter DBMS mark:'))
marks = [java, python, ai, c, dbms]
for i in range(5):
    mark = marks[i]
    if mark >= 90 and mark <= 100:
        print(subjects[i], 'A', mark)
    elif mark >= 80 and mark < 90:
        print(subjects[i], 'B', mark)
    elif mark >= 70 and mark < 80:
        print(subjects[i], 'C', mark)
    elif mark >= 60 and mark < 70:
        print(subjects[i], 'D', mark)
    elif mark >= 50 and mark < 60:
        print(subjects[i], 'E', mark)
    else:
        print(subjects[i], 'Fail', mark)
```
