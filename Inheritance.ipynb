```python
print("-----Single Inheritance---------")
class Animal:
    def A_sound(self):
        print("Animals make sounds")
class Dog(Animal):
    def D_sound(self):
        print("Dog barks")
s=Dog()
s.A_sound()
s.D_sound()
```

```python
print("--------Multiple Inheritance-----------")
class Father:
    def F_skill(self):
        print("Father knows Driving")
class Mother:
    def M_skill(self):
        print("Mother Knows Cooking")
class son(Father,Mother):
    def S_skill(self):
        print("son knows python programming")
m=son()
m.S_skill()
m.M_skill()
m.F_skill()
```

```python
print("--------Employee Salary System-----------")
class Employee:
    def basic_salary(self):
        self.salary=25000
class Bonus(Employee):
    def add_bonus(self):
        self.bonus=6000
class total_salary(Bonus):
    def total(self):
        self.basic_salary()
        self.add_bonus()
        print("Total Salary :",self.salary + self.bonus)
E=total_salary()
E.total()
```

```python
print("--------Hybrid Inheritance-----------")
class Product:
    def price(self):
        self.amount=8000
class Discount(Product):
    def discount(self):
        self.dis=300
class GST(Product):
    def gst(self):
        self.tax=190
class FinalBill(Discount, GST):
    def bill(self):
        self.price()
        self.discount()
        self.gst()
        total = self.amount - self.dis + self.tax
        print("Final Bill :",total)
b = FinalBill()
b.bill()
```
