```python
def namerow(name):
    for i in range(len(name)):
        for j in range(i+1):
            print(name[i],end='')
        print()
namerow('sachin')
```

```python
def lnamerow(name='sachin'):
    for i in range(len(name)):
        for j in range(len(name)-1,i,-1):
            print(' ',end='')
        for k in range(0,i+1):
            print(name[i],end='')
        print()
lnamerow()
```

```python
def pnamerow(name='sachim'):
    for i in range(len(name)):
        for j in range(len(name)-1,i,-1):
            print('',end=' ')
        for k in range(0,i+1):
            print(name[i],end=' ')
        print()
pnamerow(name='sachin')
```
