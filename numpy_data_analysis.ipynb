```python
import numpy as np
marks=np.array([[85,95,74],[95,58,85],[97,93,84],[74,75,94]])
total=marks.sum(axis=0)
avg=marks.mean(axis=1)
print("Total: ",total)
print("Average: ",np.round(avg,2))
print("Topper Index: ",np.argmax(total))
print("highest score: ",marks.max(axis=0))
print("Students Above 80%:",np.where(avg>80)[0])
```

```python
sales = np.array([[12000,15000,17000],[18000,21000,20000],[14000,16000,19000]])
print("Revenue: ",sales.sum(axis=1))
print("Monthly Revenue: ",sales.sum(axis=0))
print("Best Branch: ",np.argmax(sales.sum(axis=1)))
print("Highest Sales: ",sales.max())
```

```python
salary = np.array([45000,68000,72000,85000,56000])
updated = salary * 1.12
print("Updated Salary:",updated)
print("High Salary:",updated[updated > 70000])
print("Payroll:", updated.sum())
```
