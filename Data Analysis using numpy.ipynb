```python
import numpy as np
m1 = np.array([[78, 85], [92, 88]])
m2 = np.array([[2, 1], [1, 2]])
print("Matrix Multiplication using dot():")
print(np.dot(m1,m2))
```

```python
print("\nMatrix Multiplication using @:")
print(m1 @ m2)
```

```python
print("\nTranspose of Student Marks:")
print(m1.T)
```

```python
print("Random Attendance Percentage:")
print(np.random.rand(2,3))
```

```python
print("\nRandom Quiz Scores:")
print(np.random.randint(0, 10,size=(2, 3)))
```

```python
print("\nRandom Performance Distribution:")
print(np.random.randn(2, 3))
```

```python
classA = np.array([80, 75, 90])
classB = np.array([85, 88, 92])
print("Vertical Stack:")
print(np.vstack((classA, classB)))
```

```python
print("\nHorizontal Stack:")
print(np.hstack((classA, classB)))
```

```python
print("\nConcatenate:")
print(np.concatenate((classA, classB)))
```

```python
print("\nSplit Roll Numbers:")
print(np.split(np.arange(9), 3))
```

```python
scores = np.array([65, 82, 74, 90, 56, 99, 71, 88])
print("Sorted Scores:")
print(np.sort(scores))
```

```python
print("\nOriginal Index Order:")
print(np.argsort(scores))
```

```python
print("\nScores Greater Than 80:")
print(np.where(scores > 80))
```

```python
marks = np.array([35, 42, 55, 28, 76])
passed = marks >= 40
print("Pass Status:")
print(passed)
print("\nPassed Student Marks:")
print(marks[passed])
```
