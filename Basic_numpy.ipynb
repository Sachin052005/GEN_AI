```python
import numpy as np
print("---Store Student Marks in a 1D Array---")
mark=np.array([92,83,95])
print(mark)
```

```python
print("---Store Employee Monthly Sales Data---")
employee=np.array([[45000,46000,45500],[50000,55000,49000]])
print(employee)
```

```python
print("---Initialize an Empty Attendance Sheet---")
sheet=np.zeros((2,3))
print(sheet)
```

```python
print("---Analyze Dataset Properties---")
print(employee.shape)
print(employee.ndim)
print(employee.size)
print(employee.dtype)
```

```python
print("---Retrieve a Specific Record---")
print(employee[0,2])
```

```python
print("---Display the sales of all employees for the second month---")
print(employee[:,1])
```

```python
print("---Display all monthly sales of the second employee---")
print(employee[1,:])
```

```python
print("---Extract the first two rows and first two columns from the dataset---")
print(employee[0:2,0:2])
```
