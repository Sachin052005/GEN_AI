```python
print("---Deposit and withdraw money from a bank account---")
class Bank:
    def __init__(self):
        self.balance=0
    def deposite(self,amount):
        self.balance +=amount
    def withdraw(self,amount):
        if amount<=self.balance:
            self.balance -=amount
        else:
            print("Insufficient Balance")
    def cash(self):
        print("Balance:", self.balance)
B=Bank()
B.deposite(5000)
B.withdraw(1500)
B.cash()         
```

```python
print("--Find the area of a rectangle using a class--")
class rec:
    def __init__(self,l,b):
        self.l=l
        self.b=b
    def area(self):
        return self.l*self.b
r=rec(10,5)
r.area()
```

```python
print("--Calculate the area of a circle--")
class circle:
    def __init__(self,radius):
        self.radius=radius
    def cir(self):
        return 3.14*self.radius*self.radius
c=circle(7)
c.cir()
    
```

```python
print("---Create a class to issue and return books---")
class library:
    def __init__(self):
        self.books=10
    def issued(self):
        self.books -=1
    def return_book(self):
        self.books-=1
    def details(self):
        return self.books
l=library()
l.issued()
l.return_book()
l.details()
```

```python
print("---Calculate total and average marks---")
class result:
    def __init__(self,Tamil,English,Maths,Science,Social):
        self.Tamil=Tamil
        self.English=English
        self.Maths=Maths
        self.Science=Science
        self.Social=Social
    def total_Avg(self):
        total=self.Tamil+self.English+self.Maths+self.Science+self.Social
        avg=total/5
        print("Total: ",total)
        print("Avg: ",avg)
r=result(97,87,85,90,92)
r.total_Avg()
```
